```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import random
```

```python
e = torch.randn(32, 4, 68)      # 32 examples, 4 characters, 68 embeddings
emean = e.mean(0, keepdim=True) # (1, 4, 68); mean ONLY over the very first dimension
emean2 = e.mean((0,1), keepdim=True)

emean3 = e.mean(0, keepdim=False) # (4, 68); mean ONLY over the very first dimension
emean4 = e.mean((0,1), keepdim=False) # (68); mean over the first two dimensions

print(emean.shape)  # (1, 4, 68)
print(emean2.shape) # (1, 1, 68)
print(emean3.shape) # (4, 68)
print(emean4.shape) # (68)

```
